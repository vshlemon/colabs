```python
import os

filepath = "/content/MLN7496446704.mp3" # uploaded audio of this name to this path
os.path.exists(filepath) # checking its there (if not upload a file - rename as needed)
```

```python
from pyannote.audio import Pipeline, Audio
from tqdm.notebook import tqdm
import torch

pipeline = Pipeline.from_pretrained(
    'pyannote/speaker-diarization',
    use_auth_token=HUGGING_FACE_ACCESS_TOKEN
  )

pipeline.to(torch.device("cuda"))

# Load the audio file and downmix and downssample it
io = Audio(mono='downmix', sample_rate=16000)
waveform, sample_rate = io(filepath)

diarization = pipeline({"waveform": waveform, "sample_rate": sample_rate})

# print the result
for turn, _, speaker in diarization.itertracks(yield_label=True):
    print(f"start={turn.start:.1f}s stop={turn.end:.1f}s speaker_{speaker}")
```
